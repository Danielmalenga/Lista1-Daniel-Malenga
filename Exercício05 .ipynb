```python
import numpy as np
import matplotlib.pyplot as plt
N=10000
lambda1=1/28
value=4
x=np.random.uniform(0,1,N)
Xexp=-np.log(x)/lambda1
count=0
for expvalue in Xexp:
    if expvalue<=value:
        count=count+1
prob=count/N
Probana=1-np.exp(-lambda1*value)
print(Xexp)
print(prob)
print(Probana)

X=np.arange(0, 200, 0.1)
fx=lambda1*np.exp(-lambda1*X)
plt.plot(X,fx)
plt.hist(Xexp,bins=100,density=True)
plt.show()
```
