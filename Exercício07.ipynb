7) Utilizando o método da inversa gerar amostras para a distribuição
f(x)=e^x/((e^2-1)) 0<=x<=2.

```python
import numpy as np
import matplotlib.pyplot as plt

N=100000
x=np.random.uniform(0,1,N)
X= np.log((x*(np.exp(2)-1)+1))
Xa = np.arange(0, 2, 0.1)
fx=np.exp(Xa)/(np.exp(2)-1)
print(X)
plt.plot (Xa,fx)
plt.hist(X,bins=60,density=True)
plt.show()
```
