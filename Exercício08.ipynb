8) Utilizando o método da aceitação/rejeição gerar amostras para a distribuição
f(x)=1.5x^2,−1<x<1
Plotar a pdf analitica e o histograma normalizado

```python
import numpy as np
import matplotlib.pyplot as plt

n=100000
xfunction = np.array([])

for i in range(n):
    x1=np.random.uniform(-1,1)
    x2=np.random.uniform(0,1)

    while (x2 >=((1.5)*x1**2)/1.5):
        x1=np.random.uniform(-1,1)
        x2=np.random.uniform(0,1)
    xfunction = np.append(xfunction, x1)

print(xfunction)
X=np.arange(-1, 1, 0.1)
fx=1.5*X**2
plt.plot(X,fx)
plt.hist(xfunction,bins=100,density=True)
plt.show()
```
